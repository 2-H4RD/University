```python
#Практика 1
import pandas as pd
from sklearn.svm import SVC
data = pd.read_csv('svm-data.csv', header=None)
y = data.iloc[:, 0]
X = data.iloc[:, [1, 2]]
clf = SVC(kernel='linear', C=100000, random_state=241)
clf.fit(X, y)
support_numbers = sorted([i + 1 for i in clf.support_])
print(' '.join(map(str, support_numbers)))
```
